```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("lab-3//Train.csv")
plt.figure(figsize = (10,8))
plt.scatter(data['x'], data['y'], color='lightgreen')
```

```python
df = data
df = df.to_numpy()
n = df.shape[0]
m = df.shape[1]

X_train = np.zeros((n,m))
X_train[:,0] = np.ones(n)
X_train[:,1] = df[:,0]

Y_train = df[:,1]
theta = (0,0)
theta0 = []
theta1 = []
cost = []

alpha = 0.0001
itr = 100

for i in range(itr) :
    update = np.zeros(m)
    yp = np.dot(X_train,theta)
    err = yp - Y_train
    
    for j in range(m) :
        update[j] = np.sum(np.dot(err, X_train[:,j]))
        
    theta = theta - alpha/n*update
    
    theta0.append(theta[0])
    theta1.append(theta[1])
    cost.append((1/(2*n))*(np.sum(err*err)))
    
yf = theta[0] + theta[1]*X_train[:,1]

print(theta)

plt.figure(figsize = (10,8))
plt.scatter(df[:,0], df[:,1], color='lightgreen')
plt.plot(df[:,0], yf, linewidth="4")
```

```python
plt.plot(theta0)
plt.ylabel('theta-0')
plt.show()

plt.plot(theta1)
plt.ylabel('theta-1')
plt.show()

plt.plot(cost)
plt.ylabel('cost')
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
ax.set_xlabel('  Theta 0')
ax.set_ylabel('  Theta 1')
ax.set_zlabel('  Cost')
plt.plot(theta0, theta1, cost)
plt.show()
```
